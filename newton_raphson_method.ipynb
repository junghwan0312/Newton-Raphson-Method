```python
# 파이썬에서 수학 계산이나 그래프 그릴 수 있게 해주는 모듈
from sympy import *
import sympy as sp

# 변수 x, y를 수학적으로 보겠다.
x, y = sp.symbols('x y')

# fx를 x제곱(이차함수)으로 설정
fx = x**2

# 그래프를 화면에 찍는다.
plot(fx)
```

```python
# fx를 x^2-2로 설정
fx=x**2-2
# fx를 x에대해서 미분한것을 fx_prime에다 저장
fx_prime=sp.diff(fx,x)


# 함수를 출력
print("f =" , fx)
print("f' =" , fx_prime)
```

```python
# 뉴턴의 방법 구현

N = 5 # 점화식을 5번 반복
Xn= 4 # Xn은 x1, x2, ... , xn까지 변하는 변수 (초깃값(x1)을 4로 설정)

# 그래프를 화면에 한번에 출력하기 위한 설정
graphs = plot(fx,show=False)

for i in range(N):
  y = fx_prime.subs(x, Xn) * (x-Xn) + fx.subs(x,Xn) # 접선의 방정식 
  graphs.append(plot(y, line_color="red", show=False)[0]); # graphs에다 위에서 구한 접선의 방정식을 빨간색으로 저장
  print("x" + str(i+1) + "=", float(Xn)) # Xn값을 출력
  Xn = Xn - fx.subs(x, Xn) / fx_prime.subs(x, Xn) # 점화식의 x값에 Xn을 대입해서 Xn을 Xn+1로 업데이트 시켜준다
graphs.show() # 그래프를 화면에 출력
```

```python
import matplotlib.pyplot as plt
def rgb_to_hex(r, g, b):
    r, g, b = int(r), int(g), int(b)
    return '#' + hex(r)[2:].zfill(2) + hex(g)[2:].zfill(2) + hex(b)[2:].zfill(2)
```

```python
# 화면에 그래프 찍는것을 편하게 하기위해서
# drawGraph라는 함수의 형태로 만들어줌
# drawGraph에 근사해를 구하기위한 함수(fx), 초깃값(x1), 점화식을 몇번 계산할지(N)를 
# drawGraph(fx, x1, N)의 형태로 넣어주면 그래프를 찍어주고 Xn의 값을 계산해줌
def drawGraph(fx, x1, N):
  Xn = x1
  fx_prime=sp.diff(fx,x)
  graphs = plot(fx, show=False)
  for i in range(N):
    y = fx_prime.subs(x, Xn) * (x-Xn) + fx.subs(x,Xn)
    
    # 나중에 그린 그래프를 좀더 진하게 나타내주는 명령어
    graphs.append(plot(y, line_color=rgb_to_hex(255,200-i*15,200-i*15), show=False)[0]); 
    print("x" + str(i+1) + "=", Xn)
    Xn = float(Xn - fx.subs(x, Xn) / fx_prime.subs(x, Xn))
  graphs.show()
```

```python
# 3차함수
fx = (x+1)*(x-4)*(x+3)
x1= 5
N = 8


drawGraph(fx, x1, N)
```

```python
# 4차함수 
fx = 4*(x-3)*(x+2)*(x-5)*(x+6)
x1 = 7
N = 10

drawGraph(fx, x1, N)
```

```python
# 뉴터의 방법으로 파이값 구하기
fx = sp.sin(x) # fx를 sin그래프로 설정
x1 = 2 # 초깃값은 2
N = 10 # 점화식을 10번 계산

drawGraph(fx, x1, N)
```

```python
# 뉴턴의 방법으로 자연상수 구하기 
fx = sp.ln(x)-1 # fx를 ln(x)-1로 설정
x1 = 2 # 초깃값(x1)을 2로 설정
N = 10 # 점화식을 10번 계산

drawGraph(fx, x1, N)
```
